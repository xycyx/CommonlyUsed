```python
#A demo for coherence effect
#Written by Yuxuan Cheng
#11/3/2016
#yxcheng@uw.edu
#python version 2.7.12

%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
plt.style.use('ggplot')
#import pdb
from ipywidgets import interact
```

```python
def coherence(f_inter,num_wave,t_range):
    """
    simulate the independ wave and the coherence light
    f_inter is the interval of different frequency
    num_wave is the number of independent light
    t_range is the range of time (x axis)
    """
    t = np.arange(-t_range,t_range,0.001)
    plt.figure(figsize=(15,5))
    y_total = 0
    
    for i in range(0,num_wave):
        #pdb.set_trace()
        freq = f_inter*(i+30)
        phase = freq*t 
        y = np.sin(phase)  #each independent wave with different frequency
        
        y_total = y_total + y   #coherence light
        
        plt.plot(t,y)
        plt.hold(True)
    
    plt.figure(figsize=(15,5))
    plt.plot(t,y_total)
```

```python
#you can change the parameters below and observe the results
interact(coherence,f_inter=(0.0,4.0),num_wave=(0,10),t_range=10)
```
