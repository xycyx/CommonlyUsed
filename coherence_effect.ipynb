```python
#A demo for coherence effect
#Written by Yuxuan Cheng
#11/3/2016
#yxcheng@uw.edu
#python version 2.7.12

%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pdb
from ipywidgets import interact
```

```python
def coherence(f_inter,num_wave,t_range):
    """simulate the independ wave and the coherence light"""
    t = np.arange(0,t_range,0.001)
    plt.figure(figsize=(12,5))
    y_total = 0
    
    for i in range(0,num_wave):
        #pdb.set_trace()
        freq = f_inter*(i+10)
        phase = freq*t 
        y = np.sin(phase)
        
        y_total = y_total + y
        
        plt.plot(t,y)
        plt.hold(True)
    
    plt.figure(figsize=(12,5))
    plt.plot(t,y_total)
```

```python
interact(coherence,f_inter=1,num_wave=2,t_range=10);
```
